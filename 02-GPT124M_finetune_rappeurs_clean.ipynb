# Répertoires / paramètres

```python
data_rep = 'drive/MyDrive/ML/Datasets/Rappeurs/'
output_rep = 'drive/My Drive/Generated_lyrics/' # pour stockage des lyrics générées

model_folder = 'drive/My Drive/ML/gpt2/models'
checkpoint_folder = 'drive/My Drive/ML/gpt2/checkpoint'

MODEL_in = '124M_AllRapFR' # nom du modèle en entrée
```

# Import modèle rap FR

```python
sess = gpt2.start_tf_sess()

### IMPORT MODELE GPT2 FINETUNE SUR TOUT LES COUPLETS RAP_FR
gpt2.load_gpt2(sess, 
               #checkpoint_dir = model_folder, 
               model_dir = model_folder,
               model_name = MODEL_in)
```

```python
### CHECK 
gpt2.generate(sess,
              model_dir = model_folder,
              model_name = MODEL_in,
              length = 200,
              temperature = 0.95,
              top_p = 0.95,
              top_k = 40,
              prefix = "<|startoftext|>",
              #include_prefix = False,
              truncate = "<|endoftext|>",
              nsamples = 3,
              batch_size = 3)
```

# Fonction finetune artiste

```python
def finetune_on_artist(ARTISTE, STEPS):
  MODEL_out = ARTISTE + '_' + str(STEPS) # nom du modèle en sortie

  lyrics_txt = os.path.join(data_rep, ARTISTE + '.txt')
  #lyrics_npz = os.path.join(data_rep, ARTISTE + '.npz')

  # conversion pas indispensable pour un artiste
  #gpt2.encode_dataset(file_path = lyrics_txt,
  #                    model_dir = model_folder,
  #                    model_name = MODEL_in,
  #                    out_path = lyrics_npz)  
  
  #sess = gpt2.start_tf_sess()
  gpt2.finetune(sess,
                dataset = lyrics_txt, #lyrics_npz,   
                model_dir = model_folder,          
                model_name = MODEL_in,
                checkpoint_dir = checkpoint_folder,
                run_name = MODEL_out,
                steps = STEPS,
                #batch_size = 128,
                #learning_rate = 0.001,
                restore_from = 'fresh',
                print_every = 10,
                sample_every = 50,
                sample_length = 300,
                #save_every = 200,
                #overwrite = True
                reuse=True) #tf.AUTO_REUSE#
```

# Finetuning Alkpote

**Lancement**

```python
ARTISTE = 'Alkpote'
STEPS = 300
#222k tokens

MODEL_out = ARTISTE + '_' + str(STEPS) # nom du modèle en sortie
```

```python
finetune_on_artist(ARTISTE, STEPS)
```

**Tests génération sans / avec export**

```python
# "<|startoftext|>"
# graine = "j'aime"
gpt2.generate(sess,              
              checkpoint_dir = checkpoint_folder,
              run_name = 'Alkpote_300',
              length = 400,
              temperature = 0.95,
              top_p = 0.95,
              top_k = 40,
              prefix = "<|startoftext|>",#\n + graine,
              #include_prefix = False,
              truncate = "<|endoftext|>",
              nsamples = 5,
              batch_size = 5)
```

```python
#graine = ""

output_file = os.path.join(output_rep, "Alkpote_300_" + datetime.now().strftime("%m%d%Y %H:%M:%S") + ".txt")

gpt2.generate_to_file(sess,
                      destination_path = output_file,
                      checkpoint_dir = checkpoint_folder,
                      run_name = 'Alkpote_300',
                      length = 200,
                      temperature = 0.9,
                      top_p = 0.95,
                      top_k = 40,
                      prefix = "<|startoftext|>", #\n" + graine,
                      truncate = "<|endoftext|>",
                      #include_prefix = False,
                      nsamples = 5,
                      batch_size = 5)
```
