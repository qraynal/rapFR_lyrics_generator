# Export couplets en txt au format GPT-2-simple

```python
def save_couplets_to_txt(ARTISTE):

  print(ARTISTE)

  if ARTISTE == 'All':
    data_rep = 'drive/MyDrive/ML/Datasets/'
    data_file = os.path.join(data_rep, 'all_fr_verse.pkl')
    data_file_txt = os.path.join(data_rep, 'all_fr_verse.txt')

    var = 'verses' # proc_lyrics_lines / verses selon pkl

  else:
    data_rep = 'drive/MyDrive/ML/Datasets/Rappeurs/'
    data_file = os.path.join(data_rep, ARTISTE + '.pkl')
    data_file_txt = os.path.join(data_rep, ARTISTE + '.txt')

    var = 'proc_lyrics_lines' # proc_lyrics_lines / verses selon pkl

  with open(data_file, "rb") as fh:
    df = pickle.load(fh)

  print(f"\nAvec refrains : {len(df)} couplets")

  ### FILTRE SUR LES COUPLETS UNIQUEMENT
  #df_couplets = df[df['bracket'].str.contains("Couplet|couplet|NoBrackets")]
  df = df[~df['bracket'].str.contains("refrain|pont|chorus|intro|outro|transition", case = False)]
  print(f"Sans refrain|pont|chorus|intro|outro|transition : {len(df)} couplets")

  ### FILTRE SUR LES COUPLETS TROP COURTS
  df = df[df[var].map(len) > 6]
  print(f"Sans couplets < 7 lignes : {len(df)} couplets\n")

  print(f"Top 10 restant :\n{df.bracket.value_counts()[:10]}")

  ## FORMATAGE POUR GPT-2-SIMPLE : BALISES STARTOFTEXT/ENDOFTEXT POUR DELIMITER CHAQUE COUPLET
  lyrics_txt = ''
  for couplet in df[var] : 
    lyrics_txt += "<|startoftext|>\n"
    for ligne in couplet:
      lyrics_txt += ligne + "\n"
    lyrics_txt += "<|endoftext|>\n"

  with open(data_file_txt, "w") as fichier:
    fichier.write(lyrics_txt.lower())

  print(50*'-')
```

```python
ARTISTES = ["All", "Alkpote", "Kaaris", "Hamza", "Damso", "Lorenzo", "Niska", "Klub des Loosers"]

for artiste in ARTISTES:
  save_couplets_to_txt(artiste)
```
